```python
class Vocab:
  def __init__(self, csv_path):
    label= pd.read_csv(csv_path, header=0)
    idx2fine = label['소분류'].values.tolist()
    idx2middle = [x.split('-')[1] for x in label['중분류'].values.tolist()]
    idx2large = [x.split('-')[1] for x in label['대분류'].values.tolist()]


vocab = Vocab(my_dir/"notion_crawling_table.csv")
```

```python
vocab = []
for row in label.iterrows():
  r = row[1]
  item_dict = {'소분류': r['소분류'],
               '소분류_번호':  int(r['소분류 번호']),
               '중분류': r['중분류'].split('-')[1],
               '대분류': r['대분류'].split('-')[1],
               '중분류_번호': int(r['중분류'].split('-')[0]),
               '대분류_번호': int(r['대분류'].split('-')[0])
        }
  vocab.append(item_dict)

with open(f"vocab_listofdict.json", "w") as json_file:
  json.dump(vocab, json_file, ensure_ascii=False)
```

```python
title = label["소분류"][1:]
idx2title = []
title2idx = {}
for idx, category_name in enumerate(title):
    if category_name == " ":
        break  
    if '(' in category_name:
        category_name = category_name.replace('(1주차)', '')
        category_name = category_name.replace('(2주차)', '')
    category_name = category_name.strip()      
    idx2title.append(category_name)
    title2idx[category_name] = idx
```

```python
from pathlib import Path
data_dir = Path('/home/teo/userdata/nia_dataset')
wav_list = list(data_dir.rglob('*.wav'))

class_list = list(set([int(wav.stem.split('-')[2]) for wav in wav_list]))
```

```python
old_vocab =pd.read_excel("old_vocab.xlsx", header=None)
old_vocab_list = [''] + old_vocab[3].tolist()

new_keys = list(title2idx.keys())
```

```python
old_vocab_list[5]
```

```python
used_keys = [old_vocab_list[x] for x in class_list]
```

```python
used_keys
```

```python
total_candidate = {}
for wrd in used_keys:
  candidates = []
  found = False
  if wrd in title2idx:
    print(wrd)
    candidates.append(wrd)
    found = True
  else:
    for n_key in new_keys:
      if wrd in n_key:
        print(wrd, '/', n_key)
        found = True
        candidates.append(n_key)
  if not found:
    for sub_wrd in wrd.split(' '):
      if sub_wrd == '소리' or len(sub_wrd)==1:
        continue
      for n_key in new_keys:
        if sub_wrd in n_key:
          print(wrd, sub_wrd, '/', n_key)
          found = True
          candidates.append(n_key)
  if not found:
    print(f'Not found: {wrd}')
  total_candidate[wrd] = '; '.join(candidates)
```

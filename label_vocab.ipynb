```python
import torch
import pandas as pd
from pathlib import Path
```

```python
my_dir = Path('.')
label= pd.read_csv(my_dir/"notion_crawling_table.csv")
```

```python
title = label["소분류"][1:]
idx2title = {}
title2idx = {}
for idx, i in enumerate(title):
    if i == " ":
        break  
    if '(' in i:
        i = i.replace('(1주차)', '')
        i = i.replace('(2주차)', '')
    i = i.strip()      
    idx2title[idx+1] = i
    title2idx[i] = idx+1
```

```python
import json
with open(f"vocab.json", "w") as json_file:
  json.dump({'idx2title': list(title2idx), 'title2idx':title2idx}, json_file, ensure_ascii=False)
```

# Old Vocabulary


```python
from pathlib import Path
data_dir = Path('/home/teo/userdata/nia_dataset')
wav_list = list(data_dir.rglob('*.wav'))

class_list = list(set([int(wav.stem.split('-')[2]) for wav in wav_list]))
```

```python
old_vocab =pd.read_excel("old_vocab.xlsx", header=None)
old_vocab_list = [''] + old_vocab[3].tolist()

new_keys = list(title2idx.keys())
```

```python
old_vocab_list[5]
```

```python
used_keys = [old_vocab_list[x] for x in class_list]
```

```python
used_keys
```

```python
total_candidate = {}
for wrd in used_keys:
  candidates = []
  found = False
  if wrd in title2idx:
    print(wrd)
    candidates.append(wrd)
    found = True
  else:
    for n_key in new_keys:
      if wrd in n_key:
        print(wrd, '/', n_key)
        found = True
        candidates.append(n_key)
  if not found:
    for sub_wrd in wrd.split(' '):
      if sub_wrd == '소리' or len(sub_wrd)==1:
        continue
      for n_key in new_keys:
        if sub_wrd in n_key:
          print(wrd, sub_wrd, '/', n_key)
          found = True
          candidates.append(n_key)
  if not found:
    print(f'Not found: {wrd}')
  total_candidate[wrd] = '; '.join(candidates)
```

```python
import torch
import pandas as pd
from pathlib import Path
```

```python
my_dir = Path('.')
label= pd.read_csv(my_dir/"notion_crawling_table.csv")
```

```python
title = label["소분류"][1:]
idx2title = {}
title2idx = {}
for idx, i in enumerate(title):
    if i == " ":
        break  
    if '(' in i:
        i = i.replace('(1주차)', '')
        i = i.replace('(2주차)', '')
    i = i.strip()      
    idx2title[idx+1] = i
    title2idx[i] = idx+1
```

```python
import json

with open("project-1-at-2022-08-08-03-32-b8b6a32c.json", "r") as f:
  result = json.load(f)
```

```python
deleted_tags = []
confirmed_tags = []
for sample in result:
  annotation = sample['annotations'][0]
  ann_result = annotation['result']
  pred_result = annotation['prediction']['result']
  annotated_tags = [x['value']['labels'][0] for x in ann_result]
  pred_tags =  [(x['value']['labels'][0], x['value']['score']) for x in pred_result]
  
  for tag, score in pred_tags:
    if tag not in annotated_tags:
      deleted_tags.append((sample['data']['text'], tag, score))
    else:
      confirmed_tags.append((sample['data']['text'], tag, score))
confirmed_tags
```
